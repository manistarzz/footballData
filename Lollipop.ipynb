```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


#This makes sure our charts show up in our notebook
#%matplotlib inline

table=pd.read_csv("EPL.csv")

table.columns = ['Team','Rank','Points',
              'M','W','D',
              'L','G',
              'GA', 'GD','P',
              'M','W','D',
              'L','G',
              'GA', 'GD','P',
              'M','W','D',
              'L','G',
              'GA', 'GD']


table.head

print(table['P'])

#creating an empty chart
fig, axes = plt.subplots()

#These are the team colours
teamColours = ['#EF0107','#84BBFF','#EF0107','#001C58',
              '#D10022','#000000','#274488','#007FFF',
               '#FF0000','#0022FF','#670E36','#0022FF',
               '#FFD900','#FF0000','#0022FF','#660033',
               '#FFFF00','#0052CC','#FF0033','#000000']

axes.hlines(table['Team'],xmin=0,xmax=table['Points'], color=teamColours)

#drawing the markers(circle)
axes.plot(table['Points'], table['Team'], "o")
axes.set_xlim(0)

#formatting
plt.xlabel('Points')
plt.ylabel('Teams')
plt.title('EPL League Table')

#plt.hlines(y=np.arange(1,21),xmin=0,xmax=table['P'])



```
