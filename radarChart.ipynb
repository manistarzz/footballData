```python
import pandas as pd
from math import pi
import matplotlib.pyplot as plt

%matplotlib inline
#Create a data frame from Messi and Mbappe's 6 Ultimate Team data points from FIFA 23
Messi={'Pace':81,'Shooting':89,'Passing':90,'Dribbling':94,'Defending':34,'Physical':64}
Mbappe={'Pace':97,'Shooting':92,'Passing':89,'Dribbling':92,'Defending':36,'Physical':76}

data=pd.DataFrame([Messi,Mbappe], index=["Messi","Mbappe"])
data

Attributes =list(data)
AttNo = len(Attributes)

#These are the points we will draw on our radar
values = data.iloc[0].tolist()
values += values [:1]
values

#We need to find the angle between each point that our line will follow
angles = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles += angles [:1]


ax=plt.subplot(111, polar=True)

#Add labels to our axes
plt.xticks(angles[:-1],Attributes)

#Plot the line around the outside of the filled area, using the angles and values calculated before
ax.plot(angles,values)

#Fill in the area plotted in the last line
ax.fill(angles, values, 'teal', alpha=0.1)

#Give the plot a title and show it
ax.set_title("Messi")
plt.show()

#We can use the radar chart to compare the two players so lets do this for Mbappe

#Create the chart again but with both players angles and values
#We retrieve Mbappe attributes and angles
values2 = data.iloc[1].tolist()
values2 += values2 [:1]

angles2 = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles2 += angles2 [:1]


#Create the chart as before, but with both Ronaldo's and Messi's angles/values
ax = plt.subplot(111, polar=True)

plt.xticks(angles[:-1],Attributes)

ax.plot(angles,values)
ax.fill(angles, values, 'teal', alpha=0.1)

ax.plot(angles2,values2)
ax.fill(angles2, values2, 'red', alpha=0.1)

#Rather than use a title, individual text points are added
plt.figtext(0.2,0.9,"Messi",color="teal")
plt.figtext(0.2,0.85,"vs")
plt.figtext(0.2,0.8,"Mbappe",color="red")
plt.show()




```
