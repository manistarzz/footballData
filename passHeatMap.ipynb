```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mplsoccer import Pitch, VerticalPitch



%matplotlib inline

df=pd.read_csv("messiBetis.csv")

df.head()

#we make two lists one from the start and end of X locations
#The second list is the start and end for the Y location
df['x']=df['x']*1.2
df['y']=df['y']*0.8
df['endX']=df['endX']*1.2
df['endY']=df['endY']*0.8


#create the pitch
#fig,ax=plt.subplots()
#fig.set_size_inches(13.5,8)

pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(13, 5), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#22312b')

pitch.draw(ax=ax)
plt.gca().invert_yaxis()

#Create heatmap
sns.kdeplot(data=df,x="x",y="y",fill=True,shade_lowest=False,
            alpha=0.2,
            n_levels=10,
            cmap='magma')
#plt.show()







for i in range(len(df["x"])):
    if df['outcome'][i]=="Successful":
        plt.plot([int(df["x"][i]),int(df["endX"][i])],
        [int(df["y"][i]),int(df["endY"][i])],color="darkgreen")
        plt.plot(int(df["x"][i]),int(df["y"][i]),"o", color="darkgreen")

    else:
         plt.plot([int(df["x"][i]),int(df["endX"][i])],
        [int(df["y"][i]),int(df["endY"][i])],color="red")
         plt.plot(int(df["x"][i]),int(df["y"][i]),"o", color="red")

plt.xlim(0,120)
plt.ylim(0,80)

plt.title("Messi Pass Map vs Real Betis")
plt.show()


```
