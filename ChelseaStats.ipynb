```python
#importing needed libraries
import pandas as pd
import numpy as np
from math import pi
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#lets create a DataFrame using Koulibaly and Rudigers Defense stats from Fifa
Rudiger={'Interceptions':81,'Heading':81,'Def.Awareness':85,'Standing Tackle':86,'Sliding Tackle':82,"Jumping":84,"Strength":88,"Aggression":92}
Koulibaly={'Interceptions':85,'Heading':84,'Def.Awareness':90,'Standing Tackle':88,'Sliding Tackle':86,"Jumping":82,"Strength":94,"Aggression":83}

stats=pd.DataFrame([Rudiger,Koulibaly], index=["Rudiger","Koulibaly"])

Attributes =list(stats)
AttNo = len(Attributes)
stats
```

```python
#these are the points we wil draw on our radar
#These are the points we will draw on our radar
values = stats.iloc[0].tolist()
values += values [:1]
values

#We need to find the angle between each point that our line will follow
angles = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles += angles [:1]


ax=plt.subplot(111, polar=True)

#Add labels to our axes
plt.xticks(angles[:-1],Attributes)

#Plot the line around the outside of the filled area, using the angles and values calculated before
ax.plot(angles,values)

#Fill in the area plotted in the last line
ax.fill(angles, values, 'teal', alpha=0.1)

#Give the plot a title and show it
ax.set_title("Rudiger")
plt.show()
```

```python
#now lets compare koulibaly
#Create the chart again but with both players angles and values
#We retrieve Koulibaly attributes and angles
values2 = stats.iloc[1].tolist()
values2 += values2 [:1]

angles2 = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles2 += angles2 [:1]
ax = plt.subplot(111, polar=True)

plt.xticks(angles[:-1],Attributes)

ax.plot(angles,values)
ax.fill(angles, values, 'teal', alpha=0.1)

ax.plot(angles2,values2)
ax.fill(angles2, values2, 'red', alpha=0.1)

#Rather than use a title, individual text points are added
plt.figtext(0.2,0.87,"Rudiger",color="teal")
plt.figtext(0.2,0.85,"vs")
plt.figtext(0.2,0.82,"Koulibaly",color="red")
plt.show()
```

```python
#now lets compare it using a double bar graph
plt.subplots(figsize=(13,9)) # set the size that you'd like (width, height)
X=["Appearances","Clean Sheets","Goals Conceded","Blocks","Interceptions","Tackles","Clearances"]

X_axis=np.arange(len(X))

S_rudiger=[34,5,27,11,28,48,87]
S_Koulibaly=[23,5,22,13,29,45,72]

plt.bar(X_axis - 0.2, S_Koulibaly, 0.4, label = 'Koulibaly')
plt.bar(X_axis + 0.2, S_rudiger, 0.4, label = 'Rudiger')
  
plt.xticks(X_axis, X)
plt.xlabel("Statistics for the Rudiger season")
plt.ylabel("Number of Statistics")
plt.title("Statistics for the Rudiger final Season vs Koulibaly 22/23 Season")
plt.legend()
#ax.legend(fontszie=14)
plt.show()

```
